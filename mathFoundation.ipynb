```python
import numpy as np
import matplotlib.pyplot as plt
```

## Linear algebra

```python
m, n, p = 10, 5, 8
a = np.random.randn(m, n)
b = np.random.randn(n, p)
```

```python
c = a.dot(b)
```

```python
%%time
c_test = np.zeros((m, p))
for i in range(p):
    c_test[:, i] = a.dot(b[:,i])
```

```python
np.allclose(c, c_test)
```

### Large number



```python
tosses = [1000*i for i in range(300)]
heads_ratio = []
for toss in tosses:
    heads = np.random.rand(toss)
    heads_ratio.append((heads <= 0.6).mean())
```

```python
plt.plot(tosses, heads_ratio)
```
